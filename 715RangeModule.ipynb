```python
# binary search
# Goal: find the correct place to insert the new value to sorted array => binary search.
import bisect

class RangeModule(object):

    def __init__(self):
        self.X = [0, 10**9]
        self.track = [False] * 2
        

    def addRange(self, left, right, track=True):
        """
        :type left: int
        :type right: int
        :rtype: void
        """
        def index(x):
            i = bisect.bisect_left(self.X, x)
            # if the value doesn't match, that means we need to 
            # insert a new value and a new track.
            if self.X[i] != x:
                self.X.insert(i, x)
                # follows its lefthand side track because it must
                # be in the same range as its lefthand side.
                self.track.insert(i, self.track[i-1])
                print self.X
                print self.track
            return i
        i = index(left)
        j = index(right)
        
        # converge to the left
        self.X[i:j] = [left]
        self.track[i:j] = [track]
        print self.X
        print self.track
        

    def queryRange(self, left, right):
        """
        :type left: int
        :type right: int
        :rtype: bool
        """
        # bisect_right: always find the right-hand side to insert.
        i = bisect.bisect_right(self.X, left) - 1
        j = bisect.bisect_left(self.X, right)
        print i, j
        return all(self.track[i:j])
        

    def removeRange(self, left, right):
        """
        :type left: int
        :type right: int
        :rtype: void
        """
        # same idea as addRange but assign the track to False
        self.addRange(left, right, False)
        


# Your RangeModule object will be instantiated and called as such:
# obj = RangeModule()
# obj.addRange(left,right)
# param_2 = obj.queryRange(left,right)
# obj.removeRange(left,right)

obj = RangeModule()
queues = [obj.addRange,obj.addRange, obj.removeRange,obj.queryRange,obj.queryRange,obj.queryRange, obj.removeRange, obj.addRange]
params = [[10,20],[14,16],[14,16],[10,14],[13,15],[16,17], [10, 20], [10, 20]]
for fn, args in zip(queues, params):
    print fn(*args)
```
